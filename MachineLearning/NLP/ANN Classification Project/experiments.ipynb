```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, LabelEncoder
import pickle
```

```python
## Loading the dataset
data = pd.read_csv('Churn_Modelling.csv')
data
```

```python
## removing data that is not useful
data = data.drop(['Surname', 'RowNumber', 'CustomerId'], axis=1)
data
```

```python
##Using Label encoder for Female and Male and assigning 0 and 1 to them 
lb = LabelEncoder()
data['Gender'] = lb.fit_transform(data['Gender'])
```

```python
data
```

```python
from sklearn.preprocessing import OneHotEncoder
OHE = OneHotEncoder()
geo_encoder = OHE.fit_transform(data[['Geography']])
```

```python
geo_encoder
```

```python
## This is usefd to get out the different variables to which we assigned vector to
OHE.get_feature_names_out(['Geography'])
```

```python
geo_encoder_df = pd.DataFrame(geo_encoder.toarray(), columns=OHE.get_feature_names_out(['Geography']))
```

```python
geo_encoder_df
```

```python
data = pd.concat([data.drop('Geography', axis=1), geo_encoder_df], axis=1)
data
```

```python
with open('label_encoder.pkl', 'wb')as file:
    pickle.dump(lb, file)
with open('OneHotEncoder.pkl', 'wb') as file:
    pickle.dump(OHE, file)
```

```python
## Dicide the dataset into dependent and independent features
X= data.drop(['Exited'], axis=1)
Y= data['Exited']
```

### Basics
Now what we are doing here is we would make a training data and a test data we will decide how much data to to be for training and for testing using test_size parameter if test_size = say 0.2 that it means 80% training data and 20% testing data and the random state parameter is used for getting consistent result everytime we run increasing it or decreasing it just changes the way the data is split and 42 is just famous for using

```python
X_train,X_test,Y_train,Y_test = train_test_split(X,Y, test_size=0.2, random_state=42)
```

```python
X_train
```

```python
## Scaling these features
scale = StandardScaler()
X_train = scale.fit_transform(X_train)
X_test = scale.transform(X_test)
```

### Why transfrom and fit transform
so for the training data we would want to calculate the mean and variance and like that is the main data but for test we use only transform to avoid dara leakege it uses the same old parameters it does not create a new data variable 

```python
with open('scaler.pkl', 'wb' ) as file:
    pickle.dump(scale, file)
```
