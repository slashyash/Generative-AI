```python
import nltk
```

```python
sentence = """In the quiet village of Elaria, nestled between the Carpathian Mountains and a long-forgotten Roman aqueduct, lived a woman named Ishita Ray, who was both a quantum physicist and a part-time mushroom forager. Every morning, she would sip her matcha-lavender tea while checking the Schumann resonance data—believing it influenced her dreams. Her house, built from recycled airplane fuselage, had walls insulated with sheep wool and was powered entirely by piezoelectric tiles that lit up as she walked. Ishita's cat, Professor Zeno, had been taught to press a button that played Beethoven’s 5th Symphony every time it rained, which wasn’t often due to the microclimate created by the nearby bioluminescent algae farm.

One evening, while decoding a 17th-century cipher hidden in Galileo’s lesser-known notebooks, she received a call from an old friend at NASA’s Exoplanet Division, who claimed to have found microbial life signatures on Proxima b—in patterns eerily similar to Turing’s morphogenetic equations. Ishita packed her notebook, a few freeze-dried mangoes, and a relic from the Indus Valley Civilization, which she believed held some electromagnetic anomaly. She set out not just for answers but for the thrill of one more adventure, driven by curiosity, coffee, and the persistent hum of an unresolved equation."""
```

```python
words = nltk.word_tokenize(sentence)
```

```python
words
```

```python
tagged = nltk.pos_tag(words)
```

```python
tagged
```

```python
nltk.ne_chunk(tagged).draw()
```
