```python
import pandas as pd
```

```python
messages = pd.read_csv('generated_sms_messages.csv', sep= '    ', names=["label", "message"])
```

```python
messages
```

```python
import re
import nltk
```

```python
from nltk.corpus import stopwords
from nltk.stem.porter import PorterStemmer
stemmer = PorterStemmer()
```

```python
corpus =[]
for i in range(0, len(messages)):
    review = re.sub('[^a-zA-z]', ' ', messages['message'][i])
    review = review.lower().split()
    review = [stemmer.stem(word) for word in review if word not in set(stopwords.words('english'))]
    review = ' '.join(review)
    corpus.append(review)
    
```

```python
corpus
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features = 20, binary = False)
```

Now here you can max feature will decide the top 100 words and if binary = True then only 1s and 0s


```python
vec_db = cv.fit_transform(corpus).toarray()
```
