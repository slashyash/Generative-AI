```python
from tensorflow.keras.preprocessing.text import one_hot
```

```python
### sentences
sent=[  'the glass of milk',
     'the glass of juice',
     'the cup of tea',
    'I am a good boy',
     'I am a good developer',
     'understand the meaning of words',
     'your videos are good',]
```

```python
sent
```

```python
voc_size = 9000
```

```python
ohe_repr = [one_hot(words, voc_size)for words in sent]
ohe_repr
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.layers import Embedding
```

Pad sequences help in making the input of samne size by either adding 0s in starting or in end

```python
sent_length =10 
docs = pad_sequences(ohe_repr, padding='pre', maxlen=sent_length)
print(docs)
```

```python
dim = 10
```

```python
model = Sequential()
model.add(Embedding(voc_size,dim, input_length=sent_length))
model.compile(optimizer='adam', loss='mse')
```

```python
model.summary()
```

```python
model.predict(docs)
```
